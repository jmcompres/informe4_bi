```python
#%pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

# Inciso 1

#### Cargando el dataset

```python
df = pd.read_csv('datasets/healthcare-dataset-stroke-data.csv')
df
```

#### Elegimos avg_glucose_level como variable a medir

```python
ds = df[['work_type','smoking_status','avg_glucose_level']]
ds
```

#### Agrupamos por work_type y smoking_status, y calculamos la media de la variable numérica

```python
gres = ds.groupby(['work_type','smoking_status'])['avg_glucose_level'].mean().reset_index()
gres
```

#### Con los datos transformados, ya podemos empezar a trabajar en las visualizaciones y comparar

En cuanto a un gráfico de barra, podemos seguir dos estrategias, la primera es visualización con color, elemento que podemos usar para mantener la jerarquía de los datos, facilitando la comparación directa del impacto de fumar dentro de cada categoría laboral y revelando patrones estructurales. Por el contrario, la segunda opción es concatenar los atributos para aplanar las dimensiones y tratar cada combinación como una entidad única, lo cual es ideal para generar un ranking detallado de perfiles específicos de riesgo (quién tiene la glucosa más alta en términos absolutos) en lugar de comparar tendencias generales entre grupos. En este caso, se muestran las dos estrategias.

En cuanto al heatmap, es necesario crear una matriz cuyos índices y columnas sean las dimensiones.

```python
matrix = gres.pivot(index='work_type',columns='smoking_status',values='avg_glucose_level')
```

Ahora sí, creamos las visualizaciones.

```python
# Creando axes
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico 1, seaborn barplot, utilizando color para definir combinaciones
sns.barplot(data=gres, x='work_type', y='avg_glucose_level', hue='smoking_status', ax=axes[0])
axes[0].set_title("Barras y colores")
axes[0].tick_params(axis='x', rotation=45) # Se rotan las etiquetas a 45 grados para evitar solapamiento

# Gráfico 2, matplotlib barh, presentando cada combinación explícitamente
# Primero hay que preparar los datos de las combinaciones
gres_ordenado = gres.sort_values(by='avg_glucose_level', ascending=True) # Ordenamos para conveniencia del gráfico de barras horizontales
y_combination = []
x_width = []
for index, row in gres_ordenado.iterrows():
    y_combination.append(row['work_type'] + ' - ' + row['smoking_status'])
    x_width.append(row['avg_glucose_level'])
# Se dibuja
axes[1].barh(y=y_combination, width=x_width)
axes[1].set_title("Barras con combinación como etiqueta directa")

# Gráfico 3, seaborn heatmap
# El color map es ese porque mapea valores bajos a colores claros y se evitan confusiones con los nulos
sns.heatmap(matrix, annot=True, cmap="YlOrRd", ax=axes[2])
axes[2].set_title("Mapa de Calor")
axes[2].set_yticklabels(axes[2].get_yticklabels(), rotation=0) # Esto es para colocar los títulos del eje Y de manera horizontal

# Ajuste final (tight_layout ajusta automáticamente los márgenes para que los títulos y etiquetas no se superpongan entre los gráficos)
plt.tight_layout()

# Mostrar
plt.show()
```

#### Comentarios

Pensando como médicos, el heatmap con anotaciones numéricas es probablemente la opción más efectiva para comparar estos grupos. A diferencia de los gráficos de barras, que pueden generar ruido visual al comparar múltiples categorías cruzadas, el mapa de calor funciona como una matriz de riesgo intuitiva que permite capturar extremos rápidamente, ya que la intensidad del color permite un triaje visual inmediato para identificar los focos rojos, mientras que los datos numéricos explícitos satisfacen la necesidad de precisión clínica, permitiendo evaluar tanto la tendencia general como el valor exacto de cada intersección en una sola vista compacta y ordenada.

# Inciso 2

#### Cargando el dataset

```python
df2 = pd.read_csv('datasets/online_shoppers_intention.csv')
df2
```

#### Elegimos a ProductRelated_Duration como varaible numérica

Esto es la cantidad de tiempo invertida en páginas de productos similares.

```python
ds2 = df2[['ProductRelated_Duration','Revenue']]
ds2
```

#### Construímos las visualizaciones

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Visualización boxplot
sns.boxplot(x='Revenue', y='ProductRelated_Duration', data=ds2, ax=axes[0])
axes[0].set_title("Distribución de valores de ProductRelated_Duration según Revenue")

# Visualización violin plot
sns.violinplot(x='Revenue', y='ProductRelated_Duration', data=ds2, ax=axes[1])
axes[1].set_title('Comparación por Violin Plot (Densidad)')

plt.show()
```

#### Explicaciones

**Aporte de cada gráfico:**
El Boxplot resulta fundamental para visualizar estadísticas robustas y detectar anomalías. Nos permitió identificar rápidamente que la mediana del tiempo invertido por los usuarios que compran (True) es visiblemente superior a la de los que no compran. Además, destaca de manera drástica la presencia de extremos en el grupo False (puntos individuales por encima de 60,000), revelando que pasar mucho tiempo en la web no garantiza una compra. Por otro lado, el Violin Plot complementa esta visión mostrando la densidad de la distribución. Su base ancha en el grupo False confirma que la inmensa mayoría de los no compradores tienen tiempos cercanos a cero (rebote o poco interés), mientras que la forma ligeramente más estirada y elevada en el grupo True indica que la densidad de compradores se distribuye en tiempos de navegación un poco más prolongados y consistentes.

**El usuario valioso:**
Basado en la distribución observada, el usuario más valioso no es necesariamente aquel que pasa una cantidad excesiva de tiempo en el sitio (ya que los valores máximos extremos pertenecen al grupo de no compradores, posiblemente usuarios indecisos o pestañas olvidadas), sino aquel con un tiempo de navegación decente (moderado/alto) y consistente. La distribución de True para Revenue sugiere que el comportamiento de compra está correlacionado con una navegación activa y sostenida (mayor mediana y rango intercuartílico), lo que implica un interés real en los productos. Por el contrario, el grupo False muestra una polarización ineficiente, o abandonan casi de inmediato o divagan excesivamente sin convertirse en compradores. Por tanto, el equilibrio del usuario valioso se encuentra en aquellos que superan la barrera inicial de tiempo breve, demostrando interés, sin llegar a los extremos de indecisión.

Sin embargo, la tendencia no es absoluta debido a la presencia de outliers notables; se destaca particularmente un país con un nivel económico medio-alto (dimensión X con valor casi 1.5) que presenta una felicidad drásticamente baja (< 3.0) en comparación con sus pares económicos, y también hay otro otro casos de un país con bajos ingresos (dimensión X con valor casi 0) que tiene felicidad moderada.

# Inciso 4

#### Cargando el dataset

```python
df4 = pd.read_csv('datasets/WineQT.csv')
df4
```

#### Elegimos las variables fisico-químicas de pH, alcohol, acidez fija, acidez volátil, y azucar residual

```python
ds4 = df4[['pH','alcohol','fixed acidity','volatile acidity','residual sugar','quality']]
ds4
```

#### Examinamos la matriz de correlación

```python
ds4.corr()
```

#### Creamos un correlograma heatmap

```python
matriz = ds4.corr()
sns.heatmap(matriz,annot=True)
plt.show()
```

#### Comentarios

Al examinar la fila de la variable objetivo quality, se destacan dos relaciones determinantes que operan en direcciones opuestas. Primero, existe una correlación positiva moderada con el alcohol (0.48), lo que sugiere que los vinos con mayor graduación alcohólica tienden a recibir mejores puntuaciones de calidad. En segundo lugar, se observa una correlación negativa relevante con volatile acidity (-0.41); esto indica que a medida que aumenta la acidez volátil (asociada comúnmente al avinagramiento o defectos del vino), la calidad percibida disminuye significativamente.

En conclusión, según las correlaciones observadas, las variables más influyentes para predecir una mejor calidad son el contenido de alcohol (influencia positiva) y la acidez volátil (influencia negativa). El resto de las variables analizadas, como residual sugar (0.022), pH (-0.052) y fixed acidity (0.12), presentan coeficientes muy cercanos a cero, lo que implica que, por sí solas, tienen una relación lineal muy débil o casi nula con la puntuación final de calidad en este dataset.

# Inciso 5

#### Cargando el dataset

```python
df5 = pd.read_csv('datasets/meteorite-landings.csv')
df5
```

#### Obteniendo los atributos de interés

```python
df5 = df5.dropna(subset=['year','reclat','reclong','mass'])
ds5 = df5[['year','reclat','reclong','mass']]
ds5
```

#### Filtramos por año (<= 2016 porque en la fuente de los datos dice que después de esta año los datos son erróneos)

```python
ds5['year'] = ds5['year'].astype(int)
```

```python
ds5_mayor = ds5[(ds5['year'] >= 2000) & (ds5['year'] <= 2016)]
ds5_mayor
```

#### Elaboramos el mapa en base a reclat, reclong y mass

```python
gdf = gpd.GeoDataFrame(ds5_mayor,geometry=gpd.points_from_xy(ds5_mayor.reclong, ds5_mayor.reclat))
```

```python
fig, ax = plt.subplots(figsize=(18, 12))

# URL directa a un archivo .zip que contiene las fronteras de los países del mundo (dataset de Natural Earth)
url_mapa = "https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip"

# GeoPandas lee ese archivo remoto y lo convierte en un GeoDataFrame
world = gpd.read_file(url_mapa)

# color=#e0e0e0: Pinta los países de un gris claro.
# edgecolor=white: Pinta los bordes de los países de blanco.
world.plot(ax=ax, color='#e0e0e0', edgecolor='white')

# El tamaño se divide entre 500 para que no sea demasiado grande
gdf.plot(ax=ax,markersize=ds5_mayor['mass'] / 500,column='mass',cmap='magma_r',alpha=0.6,legend=True,
         legend_kwds={'label': "Masa del Meteorito (g)", 'shrink': 0.5})
# Añade una cuadrícula de latitud y longitud con líneas discontinuas y baja opacidad (0.3)
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

#### Observaciones

Observando estrictamente el mapa, se percibe un patrón geográfico muy marcado donde los impactos parecen concentrarse preferentemente en el Hemisferio Norte, formando un cinturón horizontal de actividad que atraviesa el norte de África, la Península Arábiga y Estados Unidos. Visualmente, el fenómeno sugiere una selectividad hacia ciertas masas continentales, dejando vastas áreas como los océanos, Sudamérica y gran parte del Hemisferio Sur prácticamente inmunes a estos eventos, mientras que los meteoritos de gran masa (burbujas grandes y oscuras) se presentan como anomalías aisladas y dispersas en Eurasia (exceptuando la de África a la izquierda), sin coincidir necesariamente con las zonas de mayor frecuencia de impactos menores.
